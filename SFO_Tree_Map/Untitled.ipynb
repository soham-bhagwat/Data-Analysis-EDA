```python
import pandas as pd
```

```python
df = pd.read_csv('Street_Tree_List.csv')
df.head()
```

```python
def extract_names(s):
    parts = s.split('::')
    return parts[1].strip() if len(parts) > 1 else s
```

```python
req_cols = ['TreeID','qSpecies','qCaretaker','XCoord','YCoord', 'Latitude','Longitude']
tree_df = df[req_cols]
```

```python
tree_df.isna().sum()
```

```python
top10_trees = tree_df.groupby('qSpecies').size().reset_index(name='count').sort_values(by='count', ascending=False).tail(10)
```
